```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
```

```python
service = Service("C:/Users/DELL/Downloads/chromedriver_win32/chromedriver.exe")
driver = webdriver.Chrome(service=service)
```

```python
wait=WebDriverWait(driver, 5)
```

```python
main_url='https://growjo.com/city/'
cities=['Atlanta','Austin','Boston','Houston','Dallas','San Antonio','East Bay','San Francisco','Silicon Valley','Santa Barbara',
       'Los Angeles','Orange County','San Diego','Seattle','Portland','Toronto','Boulder','Denver','Jacksonville','Tampa',
       'Orlando','Miami','Detroit','Pittsburgh','Philadelphia','Phoenix','Minneapolis','Salt Lake City','Washington DC',
       'Nashville','St Louis','New York City','Raleigh Durham','Kansas City','Charlotte']
```

```python
row_data=[]
driver.maximize_window()
for city in cities:
    links=[]
    links.append(main_url+city)
    for index in range(2,6):
        links.append(main_url+city+'/'+str(index))
    for item in links:
        driver.implicitly_wait(20)
        driver.get(item)
        while 'table' not in driver.page_source:
            time.sleep(0.1)
        soup = BeautifulSoup(driver.page_source, "html.parser")
        for tr in soup.find('table').find_all('tr'):
            row = tr.find_all('td')
            if len(row)!=8:
                continue
            else:
                link=row[7].find('a').get('href')
                data = [td.text for td in row]
                data[-1]=link
                data.insert(0,city)
                row_data.append(data)
```

```python
column_names = ["City","Rank","Company", "Industry", "Funding","Employees","Employee_Growth","Current_Jobs","Job_Links"]
df = pd.DataFrame(columns = column_names)
```

```python
index=0
for item in range(len(row_data)):
    df.loc[index,:]=row_data[item]
    index=index+1
```
