```python
i=0
for item in range(3):
    try:
        driver.get(df.loc[item,'Job_Links'])
        wait.until(EC.presence_of_element_located((By.XPATH, "//span[text()='See all jobs']")))
        driver.find_element(By.XPATH,"//span[text()='See all jobs']").click()
        wait.until(EC.presence_of_element_located((By.XPATH, "//button[text()='Date Posted']")))
        driver.find_element(By.XPATH,"//button[text()='Date Posted']").click()
        wait.until(EC.presence_of_element_located((By.XPATH, "//span[text()='Past Month']")))
        driver.find_element(By.XPATH,"//span[text()='Past Month']").click()
    
        staleElement=True
        while staleElement:
            try:
                elements=driver.find_elements(By.XPATH,"//div[contains(@class,'reusable-search-filters-buttons')]/button[contains(@data-control-name,'filter_show_results')]/span");
                elements[1].click()
                staleElement = False
            except:
                staleElement = True
    
        time.sleep(15)
        wait.until(EC.presence_of_element_located((By.XPATH, "//li[contains(@class,'jobs-search-results')]")))
        jobs_list=driver.find_elements(By.XPATH,"//li[contains(@class,'jobs-search-results')]")
        
        wait.until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class,'artdeco-entity-lockup__title')]//a[contains(@href,'jobs')]")))
        jobs_url=driver.find_elements(By.XPATH,"//div[contains(@class,'artdeco-entity-lockup__title')]//a[contains(@href,'jobs')]")
        
        print(len(jobs_url))

        for value in range(len(jobs_list)):
            lst1=list(df_list[item])
            lst1.append(jobs_url[value].get_attribute('href'))
            try:
                df1.loc[i,:]=lst1
                i=i+1
            except:
                continue 
    except:
        df_list[item][-1]='No jobs yet'
        df1.loc[i,:]=df_list[item]
        print("No jobs yet")
        i=i+1
```
